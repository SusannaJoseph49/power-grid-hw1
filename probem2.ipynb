```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
file_path = 'co2GDP_2015.xlsx'
df = pd.read_excel(file_path)

square_marker_countries = [
    "United Arab Emirates", "Kuwait", "Australia", "Bahrain", "United States", 
    "Canada", "Luxembourg", "New Zealand", "Ireland", "Estonia", "Saudi Arabia",
    "Singapore", "Belgium", "Netherlands", "Finland", "Denmark", "Norway",
    "Germany", "Israel", "United Kingdom", "Japan", "Slovenia", "Spain", 
    "France", "Sweden", "Switzerland", "Hong Kong"
]

circle_marker_countries = [
    "Trinidad", "Turkmenistan", "Czech Republic", "Russia", "Oman", "Kazakhstan", 
    "Greece", "South Korea", "Cyprus", "Austria", "Iceland", "Venezuela", 
    "Poland", "Slovakia", "Ukraine", "South Africa", "Belarus", "Argentina", 
    "Uruguay", "Portugal", "Hungary", "Bulgaria", "Croatia", "Latvia", 
    "Lithuania", "Malaysia", "Turkey", "Chile", "Romania", "Brazil", "China", 
    "Mexico", "Panama", "Costa Rica", "Bosnia", "Macedonia", "Albania", "India"
]

df_filtered_squares = df[df['Entity'].isin(square_marker_countries)]
df_filtered_circles = df[df['Entity'].isin(circle_marker_countries)]

plt.figure(figsize=(12, 8))
plt.scatter(df_filtered_squares['GDP per capita'], df_filtered_squares['Annual greenhouse gas emissions in CO₂ equivalents/per person'], 
            color='blue', marker='s', s=50, label='Square Markers')

plt.scatter(df_filtered_circles['GDP per capita'], df_filtered_circles['Annual greenhouse gas emissions in CO₂ equivalents/per person'], 
            color='green', marker='o', s=50, label='Circle Markers')

for i, row in df_filtered_squares.iterrows():
    plt.text(row['GDP per capita'], row['Annual greenhouse gas emissions in CO₂ equivalents/per person'], row['Entity'], fontsize=8, ha='right')

for i, row in df_filtered_circles.iterrows():
    plt.text(row['GDP per capita'], row['Annual greenhouse gas emissions in CO₂ equivalents/per person'], row['Entity'], fontsize=8, ha='right')

plt.xlabel('GDP per capita')
plt.ylabel('Annual greenhouse gas emissions in CO₂ equivalents/per person')
plt.title('Annual Greenhouse Gas Emissions vs GDP per Capita with Different Markers')
plt.legend()
plt.grid(True)
plt.show()
```
