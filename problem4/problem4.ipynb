```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
data = {
    "Small Hydro Power": [4849, 4944],  
    "Wind Power": [40358, 42633],
    "Bio Power & Waste to Energy": [10682, 10802],
    "Solar Power": [53997, 66780]
}
data_in_gw = {key: [val / 1000 for val in value] for key, value in data.items()}
categories = list(data_in_gw.keys())
years = ["2021-22", "2022-23"]
bar_width = 0.35
index = np.arange(len(categories))
fig, ax = plt.subplots(figsize=(10, 6))
bar_2021_22 = ax.bar(index, [data_in_gw[cat][0] for cat in categories], bar_width, label='2021-22', color='b')
bar_2022_23 = ax.bar(index + bar_width, [data_in_gw[cat][1] for cat in categories], bar_width, label='2022-23', color='g')
for bars in [bar_2021_22, bar_2022_23]:
    for bar in bars:
        yval = bar.get_height()
        ax.text(bar.get_x() + bar.get_width() / 2, yval, f'{yval:.2f}', ha='center', va='bottom')

ax.set_xlabel('Renewable Energy Type')
ax.set_ylabel('Installed Capacity (GW)')
ax.set_title('Installed Capacity of Grid-Interactive Renewable Power\nDuring 2021-22 and 2022-23')
ax.set_xticks(index + bar_width / 2)
ax.set_xticklabels(categories)
ax.legend()
plt.tight_layout()
plt.show()
```

```python
data = {
    "Small Hydro Power": [4849, 4944], 
    "Wind Power": [40358, 42633],
    "Bio Power & Waste to Energy": [10682, 10802],
    "Solar Power": [53997, 66780]
}

data_in_gw = {key: [val / 1000 for val in value] for key, value in data.items()}
percentage_change = {key: ((data_in_gw[key][1] - data_in_gw[key][0]) / data_in_gw[key][0]) * 100 for key in data_in_gw}

categories = list(data_in_gw.keys())
years = ["2021-22", "2022-23"]
bar_width = 0.3
index = np.arange(len(categories))
fig, ax1 = plt.subplots(figsize=(12, 6))

bar_2021_22 = ax1.bar(index, [data_in_gw[cat][0] for cat in categories], bar_width, label='2021-22 (GW)', color='b')
bar_2022_23 = ax1.bar(index + bar_width, [data_in_gw[cat][1] for cat in categories], bar_width, label='2022-23 (GW)', color='g')
bar_percentage = ax1.bar(index + 2 * bar_width, [percentage_change[cat] for cat in categories], bar_width, label='Percentage Change (%)', color='r', alpha=0.7)
for bars in [bar_2021_22, bar_2022_23]:
    for bar in bars:
        yval = bar.get_height()
        ax1.text(bar.get_x() + bar.get_width() / 2, yval, f'{yval:.2f}', ha='center', va='bottom')

for bar in bar_percentage:
    yval = bar.get_height()
    ax1.text(bar.get_x() + bar.get_width() / 2, yval, f'{yval:.2f}%', ha='center', va='bottom')

ax1.set_xlabel('Renewable Energy Type')
ax1.set_ylabel('Installed Capacity (GW)')
ax1.set_title('Installed Capacity of Grid-Interactive Renewable Power\nDuring 2021-22 and 2022-23 (GW) and Year-on-Year Percentage Change')
ax1.set_xticks(index + bar_width)
ax1.set_xticklabels(categories)
ax1.legend()
plt.tight_layout()
plt.show()
```
